DATAPREPROCESSING FOR TWEETS

```python
# Initialize lists to store tweet details
created_at = []
tweet_id = []
text = []
user_id = []
tweet_country = []
tweet_language = []
entities_hashtag = []
entities_usermentions = []
date_stamp=[]

# Loop through each entry in the DataFrame
for i in range(0, len(Corona_df_combined)):
    if Corona_df_combined['id_str'][i] and not Corona_df_combined["text"][i].startswith('RT'):
        # Since the tweet is not a retweet, collect its data

        # Convert the string to a datetime object
        date_time_obj = pd.to_datetime(Corona_df_combined["created_at"][i])

        created_at.append(Corona_df_combined["created_at"][i])
        date_stamp.append(date_time_obj.strftime("%Y-%m-%d"))
        tweet_id.append(Corona_df_combined["id_str"][i])
        text.append(Corona_df_combined["text"][i])
        user_id.append(Corona_df_combined["user"][i]['id'])
        tweet_language.append(Corona_df_combined["lang"][i])
        
        # Append country or None if not available
        if Corona_df_combined["place"][i] and 'country' in Corona_df_combined["place"][i]:
            tweet_country.append(Corona_df_combined["place"][i]['country'])
        else:
            tweet_country.append(None)
        
        # Handle user mentions
        if 'user_mentions' in Corona_df_combined["entities"][i]:
            temp_user_mentions = [mention['id'] for mention in Corona_df_combined["entities"][i]['user_mentions']]
            entities_usermentions.append(temp_user_mentions)
        else:
            entities_usermentions.append(None)  # Append None if no user mentions are found

        # Handle hashtags
        if 'hashtags' in Corona_df_combined["entities"][i]:
            temp_hashtags = [hashtag['text'] for hashtag in Corona_df_combined["entities"][i]['hashtags']]
            entities_hashtag.append(temp_hashtags)
        else:
            entities_hashtag.append(None)  # Append None if no hashtags are found

# Create the DataFrame
Tweet_df = pd.DataFrame({
    'CREATED_AT': created_at,
    'DATE_STAMP': date_stamp,
    'TWEET_ID': tweet_id,
    'TEXT': text,
    'USER_ID' : user_id,
    'TWEET_LANGUAGE' : tweet_language,
    'TWEET_COUNTRY' : tweet_country,
    'ENTITIES_HASHTAGS': entities_hashtag,
    'ENTITIES_USER_MENTIONS': entities_usermentions,
})

# Additional processing of hashtags and user mentions
Tweet_df['ENTITIES_HASHTAGS'] = Tweet_df['ENTITIES_HASHTAGS'].apply(lambda x: [str(tag) for tag in set(x)] if isinstance(x, list) and x else None)
Tweet_df['ENTITIES_USER_MENTIONS'] = Tweet_df['ENTITIES_USER_MENTIONS'].apply(lambda x: ["'" + str(mention) + "'" for mention in set(x)] if isinstance(x, list) and x else None)
```
